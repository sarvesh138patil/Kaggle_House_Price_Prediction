```python
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error
from sklearn.model_selection import train_test_split
```

```python
home_data = pd.read_csv("/kaggle/input/home-data-for-ml-course/train.csv")
```

```python
home_data.head()
```

```python
y = home_data.SalePrice
```

```python
features = ['LotArea', 'YearBuilt', '1stFlrSF', '2ndFlrSF', 'FullBath', 'BedroomAbvGr', 'TotRmsAbvGrd']
```

```python
X = home_data[features]
```

```python
X.head()
```

```python
X_train, X_val, y_train, y_val = train_test_split(X, y, random_state=1)
```

```python
rf_model = RandomForestRegressor(random_state=1)
rf_model.fit(X_train, y_train)
```

```python
rf_val_prediction = rf_model.predict(X_val)
```

```python
rf_val_mae = mean_absolute_error(rf_val_prediction, y_val)
rf_val_rmse = np.sqrt(mean_squared_error(rf_val_prediction, y_val))
```

```python
print("Validation MAE for Random Forest Model: {:,.0f}".format(rf_val_mae))
print("Validation MAE for Random Forest Model: {:,.0f}".format(rf_val_rmse))
```

```python
rf_model_on_full_data = RandomForestRegressor()
```

```python
rf_model_on_full_data.fit(X, y)
```

```python
test_data_path = "/kaggle/input/home-data-for-ml-course/test.csv"
```

```python
test_data = pd.read_csv(test_data_path)
```

```python
X_test = test_data[features]
```

```python
test_preds = rf_model_on_full_data.predict(X_test)
```
